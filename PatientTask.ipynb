```python
#Merge the patient and encounter DataFrames on the "Patient_id" column 
#to include demographic information for all encounters.
merge_pe=pd.merge(df_encounter, df_patient, on='patient_id',suffixes=('_left', '_right'))
print(merge_pe)
```

```python
#check data types of column
merge_pe.dtypes
```

```python
# Convert date columns to date format
merge_pe['start_date'] = pd.to_datetime(merge_pe['start_date'], format='%Y%m%d')
merge_pe['end_date'] = pd.to_datetime(merge_pe['end_date'], format='%Y%m%d')
print(merge_pe)
```

```python
#Filter the merged DataFrame to include only 
#encounters that occurred in 2019.
merge_pe_2019 = merge_pe[merge_pe['start_date'].dt.year == 2019]
print(merge_pe_2019)
```

```python
#Merge the patient and encounter DataFrames to diagonsis df.
merge_ped=pd.merge(merge_pe_2019, df_diagnosis, on='encounter_id',suffixes=('_left', '_right'))
print(merge_ped)
```

```python
#Now we group the filtered DataFrame by "Patient_id" and select 
#the most recent encounter for each patient in 2019.
most_recent_encounter = merge_ped.loc[merge_ped.groupby('patient_id_left')['start_date'].idxmax()]
print(most_recent_encounter)
```

```python
most_recent_encounter.dtypes
```

```python
# Create dummy variables for each unique code
dummy_df = pd.get_dummies(most_recent_encounter['code'])

# Concatenate the dummy variables with the original DataFrame
pivoted_df = pd.concat([most_recent_encounter[['encounter_id', 'patient_id_left']], dummy_df], axis=1)

# Group by 'encounter_id' and 'patient_id_left' and aggregate the dummy variables
pivoted_df = pivoted_df.groupby(['encounter_id', 'patient_id_left']).max()

print(pivoted_df)
```

```python

# Select required columns from the original DataFrame
final_df = most_recent_encounter[['patient_id_left', 'sex', 'year_of_birth', 'encounter_id', 'start_date']]

# Merge with the pivoted DataFrame
final_df = final_df.merge(pivoted_df, on=['encounter_id', 'patient_id_left'], how='left')

print(final_df)


```

```python
final_df.to_csv("C:/Rahul/USF/Archive", index=False)
```
